```python
teams={'Kolkata knight riders':{'player':[],'purse':32.7,'purseUsed':0},
          'Sunrise Hyderabad':{'player':[],'purse':34.00,'purseUsed':0},
          'Chennai super kings':{'player':[],'purse':31.40,'purseUsed':0},
          'Gujarat Titans':{'player':[],'purse':38.15,'purseUsed':0},
          'Punjab kings':{'player':[],'purse':29.10,'purseUsed':0}}
def add_player(team_name,player_name,selling_price):
    if team_name in teams and teams[team_name]['purse']>=selling_price:
        teams[team_name]['player'].append({'player_name':player_name,'selling_price':selling_price})
        teams[team_name]['purse'] -=selling_price
        teams[team_name]['purseUsed']+=selling_price
        print(f"{player_name} added to {team_name} for {selling_price : 0.2f} Cr.")
    else:
        print(f"Error : Insufficient funds for {team_name} or invalid team name")
def sort_teams():
    sorted_teams=sorted(teams.items(),key= lambda x: x[1]['purseUsed'],reverse=True)
    return sorted_teams
def display():
    for team_name,details in teams.items():
        print(f'''{team_name} : Purse remaining = {details['purse']:0.2f} Cr.
                    Purse Used = {details['purseUsed']:0.2f} Cr.''')
        print("Players : ")
        for player in details['player']:
            print(f'''    {player['player_name']} --> {player['selling_price']} Cr.''')
        print()
```

```python
#   Design a Train Ticket Booking System
#   Consider a scenario where passengers need to book train tickets for a one-way journey or a
#   return journey. Design a Python program using basic functions to implement a Train Ticket
#   Booking System. The system should allow users to perform the following tasks:
#   1. Book One-Way Ticket:
#   • Passengers should be able to select the source and destination stations.
#   • Display available trains between the selected stations.
#   • Allow passengers to choose a train, enter the date of the journey, and select the
#   coach type (Sleeper/3AC/2AC/1AC). Prices will vary according to the coach type.
#   • Display the passenger's name, train details, journey date, coach type, PNR, and
#   ticket price.
#   2. Get Trip Info:
#   • Passengers should be able to retrieve information about their booked trip using the
#   PNR number.
#   • Display the train details, source, and destination of the booked trip.
#   3. Book Return Journey:
#   • If a passenger has booked a one-way ticket, provide an option to book a return
#   journey.
#   • Automatically swap the source and destination stations for the return journey.
#   • Display available trains for the return journey.
#   • Allow passengers to choose a return train, enter the return date, and select the coach
#   type.
#   • Generate a return ticket with a new PNR number.
#   4. Exit:
#   • Allow users to exit the program.
#   Requirements:
#   • Use functions to organize the code logically.
#   • Ensure proper error handling for invalid inputs.
#   • Include sample data for at least four trains with different routes and coach prices.
#   • Write a Python program that fulfills the requirements mentioned above. Include
#   comments to explain the logic and structure of your code. Test your program with
#   various scenarios to ensure its correctness. 

def make_train(name, source, destination, coach_prices):
    return {
        "name": name,
        "source": source,
        "destination": destination,
        "coach_prices": coach_prices,
        "passengers": []
    }

def display_menu():
    print("1. Book One-Way Ticket")
    print("2. Get Trip Info")
    print("3. Book Return Journey")
    print("4. Exit")

def get_available_trains(trains, source, destination):
    return [t for t in trains if t["source"] == source and t["destination"] == destination]

def swap_stations(source, destination):
    return destination, source

def is_pnr_valid(trains, pnr):
    for t in trains:
        for p in t["passengers"]:
            if p.get("ticket") and p["ticket"].get("pnr") == pnr:
                return False
    return True

def make_ticket(train, date, coach_type, pnr, kind="oneway"):
    return {
        "train": train,
        "date": date,
        "coach_type": coach_type,
        "pnr": pnr,
        "kind": kind
    }

def book_one_way_ticket(trains, passenger, train, date, coach_type, user_provided_pnr):
    if not is_pnr_valid(trains, user_provided_pnr):
        print("Invalid PNR. Please choose a unique PNR.")
        return None

    if coach_type not in train["coach_prices"]:
        print("Invalid coach type.")
        return None

    ticket = make_ticket(train, date, coach_type, user_provided_pnr, kind="oneway")
    passenger["ticket"] = ticket
    train["passengers"].append({"name": passenger["name"], "ticket": ticket})
    return ticket

def book_return_ticket(trains, passenger, train, date, coach_type, user_provided_pnr):
    if not is_pnr_valid(trains, user_provided_pnr):
        print("Invalid PNR. Please choose a unique PNR.")
        return None

    if coach_type not in train["coach_prices"]:
        print("Invalid coach type.")
        return None

    ticket = make_ticket(train, date, coach_type, user_provided_pnr, kind="return")
    passenger["ticket"] = ticket
    train["passengers"].append({"name": passenger["name"], "ticket": ticket})
    return ticket

def get_trip_info(trains, pnr):
    for t in trains:
        for p in t["passengers"]:
            if p.get("ticket") and p["ticket"].get("pnr") == pnr:
                return t
    return None

trains_data = [
    make_train("Shatabdi Express(12010)", "Ahmedabad", "Mumbai", {"Sleeper": 700, "3AC": 1300, "2AC": 1800, "1AC": 2300}),
    make_train("Duranto Express(12268)",  "Ahmedabad", "Mumbai", {"Sleeper": 600, "3AC": 1100, "2AC": 1600, "1AC": 2100}),
    make_train("Shatabdi Express(12009)", "Mumbai", "Ahmedabad", {"Sleeper": 700, "3AC": 1300, "2AC": 1800, "1AC": 2300}),
    make_train("Duranto Express(12267)",  "Mumbai", "Ahmedabad", {"Sleeper": 600, "3AC": 1100, "2AC": 1600, "1AC": 2100}),
]

passenger_name = input("Enter your name: ")
passenger = {"name": passenger_name, "ticket": None}

while True:
    display_menu()
    choice = input("Enter your choice (1-4): ").strip()

    if choice == "1":
        source = input("Enter source station: ").strip()
        destination = input("Enter destination station: ").strip()
        available = get_available_trains(trains_data, source, destination)

        if not available:
            print("No trains available for the selected route.")
            continue

        print("Available Trains:")
        for i, tr in enumerate(available, 1):
            print(f"{i}. {tr['name']} - {tr['source']} to {tr['destination']}")

        try:
            idx = int(input("Select a train (enter the number): ").strip()) - 1
            selected_train = available[idx]
        except (ValueError, IndexError):
            print("Invalid selection.")
            continue

        date_of_journey = input("Enter date of journey (YYYY-MM-DD): ").strip()
        selected_coach_type = input("Select coach type (Sleeper/3AC/2AC/1AC): ").strip()

        user_pnr = input("Enter your preferred PNR: ").strip()
        ticket = book_one_way_ticket(trains_data, passenger, selected_train, date_of_journey, selected_coach_type, user_pnr)

        if ticket:
            print("One-way Ticket booked successfully!")
            print(f"PNR: {ticket['pnr']}")
            print(f"Passenger: {passenger['name']}")
            print(f"Train: {selected_train['name']}")
            print(f"Date: {ticket['date']}")
            print(f"Coach Type: {ticket['coach_type']}")
            print(f"Ticket Price: {selected_train['coach_prices'][selected_coach_type]}")

    elif choice == "2":
        pnr = input("Enter your PNR: ").strip()
        info = get_trip_info(trains_data, pnr)
        if info:
            print(f"Train: {info['name']}\nSource: {info['source']}\nDestination: {info['destination']}")
        else:
            print("Invalid PNR. Please check and try again.")

    elif choice == "3":
        if not passenger.get("ticket"):
            print("You need to book a one-way ticket first.")
            continue

        last_train = passenger["ticket"]["train"]
        ret_src, ret_dst = swap_stations(last_train["source"], last_train["destination"])
        ret_trains = get_available_trains(trains_data, ret_src, ret_dst)

        if not ret_trains:
            print("No trains available for the return journey.")
            continue

        print("Available Trains for Return Journey:")
        for i, tr in enumerate(ret_trains, 1):
            print(f"{i}. {tr['name']} - {tr['source']} to {tr['destination']}")

        try:
            idx = int(input("Select a return train (enter the number): ").strip()) - 1
            selected_return_train = ret_trains[idx]
        except (ValueError, IndexError):
            print("Invalid selection.")
            continue

        return_date = input("Enter return date (YYYY-MM-DD): ").strip()
        selected_return_coach_type = input("Select return coach type (Sleeper/3AC/2AC/1AC): ").strip()

        user_return_pnr = input("Enter your preferred PNR for return journey: ").strip()
        ret_ticket = book_return_ticket(trains_data, passenger, selected_return_train, return_date, selected_return_coach_type, user_return_pnr)

        if ret_ticket:
            print("Return ticket booked successfully!")
            print(f"Return PNR: {ret_ticket['pnr']}")

    elif choice == "4":
        print("Exiting the program. Thank you!")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```
