```python
# wap to fetch any character from user and print all its relative types
word=input("Enter any thing : ")
if word.isalnum():
    print("It is alphanumeric ")
    if word.isalpha():
        print("It is alphabet")
        if word.isupper():
            print("It is uppercase")
        else:
            print("It is lowercase")
    else :
        print("It is a Digit")
elif word.isspace():
    print("It is a space")
else:
    print("It is non-space special character")
```

```python
# WAP to check all the validation OF USER defied password
# 1) it should have minimum 8 characters and maximum 20
# 2) it should have 1 minimum uppercase alphabet and lowercase alphabet and 1 digit and should not have any space
# 3) it should 1 special characters "#@$" one from these and last show the user's password id valid or not
# and if not then show all imrovement
password=input("Enter any Password :")
check_lower=False
check_upper=False
check_space=True
Check_special=False
check_digit=False
check_length=len(password)<8 or len(password)>20
for i in password:
    if password.isupper()==False:
        check_upper=True
    if password.islower()==False:
        check_lower=True
    if password.isspace():
        check_space=False
    if password.isnumeric()==False:
        check_digit=True
    if password not in '@#$':
        Check_special=True
if check_length:
    print("length should be minimum 8 characters and maximum 20 characters")
if check_upper==False:
    print("please add an Uppercase character in Password")
if check_lower==False:
    print("please add an Lowercase character in Password")
if check_digit==False:
    print("please add an Digit character in Password")
if check_space==False:
    print("Password can't have space")
if Check_special==False:
    print("please add an special character (@,#,$) in Password")
if(check_length and check_upper and check_digit and check_lower and check_space==False and Check_special):
    print("It is a valid password")
```

```python
# WAP to reverse words in a sentence keeping sequences of word as it is
IN=input("ENTER A SENTENCE : ")
result=""
IN=IN.split(" ")
for i in IN:
    OUT=i[::-1]
    result+=OUT+" "
print(result)
```

```python
String=input("Enter code")
key=int(input("Enter key"))
length=len(String)
part=length//key
extra=length%key
part1=String[:(part+1)*extra]
part2=String[(part+1)*extra:]
mes=""
for i in range (part+1):
    if i<part:
        mes+=part1[i::part+1]+part2[i::part]
    else:
        mes+= part1[i::part+1]
print(mes)
```
